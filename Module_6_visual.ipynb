```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv(r"C:\Users\kings\OneDrive\Desktop\Python\user_data.users.csv")

# Visualization 1: Ages with the highest income
plt.figure(figsize=(10, 6))
plt.bar(df['age'], df['total_income'], color='blue')
plt.title('Ages with the Highest Income')
plt.xlabel('Age')
plt.ylabel('Total Income')
plt.savefig('highest_income_by_age.png')
plt.show()


# Visualization 2: Gender distribution across spending categories
gender_expenses = df.groupby('gender').sum()
gender_expenses.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Gender Distribution Across Spending Categories')
plt.xlabel('Gender')
plt.ylabel('Total Expenses')
plt.savefig('gender_distribution_expenses.png')
plt.show()
```
